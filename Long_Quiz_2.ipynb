```python
#Implement the graph using Python dictionaries. This must display all the vertices and edges of the graph
class graph:
   def __init__(self,ggraph=None):
      if ggraph is None:
         ggraph = {}
      self.ggraph = ggraph

   def edges(self):
      return self.findedges()
   def findedges(self):
      edgename = []
      for vrtx in self.ggraph:
         for nxtvrtx in self.ggraph[vrtx]:
            if {nxtvrtx, vrtx} not in edgename:
               edgename.append({vrtx, nxtvrtx})
      return edgename
graph_elements = {
   "e": ["i","j","k","m"],
   "f" : ["i", "j","k","p"],
   "h" : ["i","j","k"],
   "i" : ["e","h","f"],
   "j" : ["e","f","h"],
   "k" : ["e","f","h"],
   "m" : ["e"],
   "p" : ["f"],

}
g = graph(graph_elements)
print(g.edges())
```
